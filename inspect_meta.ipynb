```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('./meta.csv')

# Counting how many times each transformation was applied
transformation_counts = df.sum()

# Normalize the counts to make the maximum value equal to the number of rows
total_rows = len(df)
print(total_rows)

# Plotting the bar plot
plt.figure(figsize=(10, 6))
transformation_counts.plot(kind='bar', color='skyblue')
plt.title("Scaled Frequency of Transformations Applied")
plt.xlabel("Transformation")
plt.ylabel("Scaled Count")
plt.xticks(rotation=45, ha='right')
plt.ylim(0, total_rows)  # Set the y-axis limit to the number of rows
plt.tight_layout()

# Display the plot
plt.show()
```
